# Set the rules/functions by regular expression

I set 4 functions, you have to set them based on your design

This code defines four functions, each with the purpose of removing specific types of characters or content from a given text.



---

**remove_non_english(text):**

This function aims to remove non-English characters from the text.
It uses a regular expression re.sub('[^\x00-\x7F]', '', str(text)) to match and remove all characters that are not within the ASCII encoding range (i.e., non-English characters).
It returns the processed text with all non-English characters removed.


---


**remove_hex_characters(text):**

This function's purpose is to remove hexadecimal characters (characters starting with \x) from the text.
It uses a regular expression re.sub(r'(\\x[0-9A-Fa-f]+)', '', str(text)) to match and remove all hexadecimal characters starting with \x.
It returns the processed text with all hexadecimal characters removed.


---


**remove_html_codes(text):**

This function aims to remove HTML tags and their content from the text.
It uses a regular expression re.sub('<.*?>', '', str(text)) to match and remove all HTML tags.
It returns the processed text with all HTML tags and their content removed.


---


**remove_pure_links(text):**

This function's purpose is to remove web links (URLs) from the text.
It uses a regular expression re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', str(text)) to match and remove all web links.
It returns the processed text with all web links removed.


---


These functions are typically used during text preprocessing to clean or standardize text data. Depending on your needs, you can choose to use these functions to clean text data for further analysis or processing.


```python

def remove_non_english(text):
    return re.sub('[^\x00-\x7F]', '', str(text))

def remove_hex_characters(text):
    return re.sub(r'(\\x[0-9A-Fa-f]+)', '', str(text))

def remove_html_codes(text):
    return re.sub('<.*?>', '', str(text))

def remove_pure_links(text):
    return re.sub(r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', str(text))


```

# Clean the dataset

```python
df = pd.read_csv('/content/drive/My Drive/30016/Phishing_Email.csv')

df['Email Text'] = df['Email Text'].apply(remove_non_english)
df['Email Text'] = df['Email Text'].apply(remove_hex_characters)
df['Email Text'] = df['Email Text'].apply(remove_html_codes)
df['Email Text'] = df['Email Text'].apply(remove_pure_links)


df.to_csv('/content/drive/My Drive/30016/cleaned_file.csv', index=False)
```
